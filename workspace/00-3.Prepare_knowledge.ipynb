```python
from pypdf import PdfReader
import json
import re
```

```python
pdf_path = "/home/jovyan/project/data/Guidebook_전자부품(배터리팩) 품질보증 AI 데이터셋.pdf"
```

```python
def extract_knowledge_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    knowledge_base = []
    
    # 가이드북 분석 결과, 핵심 지식은 7, 8, 9페이지에 집중되어 있음 
    # pypdf는 0부터 시작하므로 index 6, 7, 8이 해당됨
    target_pages = [6, 7, 8]
    
    for page_num in target_pages:
        page = reader.pages[page_num]
        text = page.extract_text()
        
        # 1. 텍스트 정제 (줄바꿈 및 불필요한 공백 제거)
        clean_text = re.sub(r'\s+', ' ', text).strip()
        
        # 2. 불량 유형별로 섹션 나누기 (간단한 규칙 기반 청킹)
        # 가이드북의 주요 키워드: 용량불량, 용접불량, 센싱와이어불량, 센서불량 
        sections = re.split(r'-(?=\s*[가-힣]+불량)', clean_text)
        
        for section in sections:
            if len(section) > 50: # 의미 있는 길이의 텍스트만 저장
                knowledge_base.append({
                    "page": page_num + 1,
                    "content": section.strip()
                })

    # 3. JSON으로 저장 (BM25 엔진의 소스로 사용)
    with open('/home/jovyan/project/data/battery_knowledge.json', 'w', encoding='utf-8') as f:
        json.dump(knowledge_base, f, ensure_ascii=False, indent=4)
        
    print(f"✅ 지식 베이스 구축 완료: battery_knowledge.json")
    print(f"📊 총 {len(knowledge_base)}개의 지식 섹션이 추출되었습니다.")

# 실행
if __name__ == "__main__":
    extract_knowledge_from_pdf(pdf_path)
```
